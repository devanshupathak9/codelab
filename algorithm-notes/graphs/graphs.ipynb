## **Disjoint Set (Union - Find)**

**Union - Find**, also knows as Disjoint set union, is a data structure used to track and merge non-overlapping (disjoint set).

It supports 2 operations:
1. **Find(x)**: It is used to find the set to which element x belongs to.
2. **Union(x, y)**: It is used to merge set containing x and y.

**Optimizations**:
1. **Path Compression**: While finding the parent of a node, point every on the way directly to the root. This will reduce the search time to O(1).

2. **Rank (Union by Rank)**: Rank is the estimate hieght of a tree. Always keeps smaller under a higher tree.

```python
class disjoint_set:
    def __init__(self, n):
        self.parents = [i for i in range(n + 1)]

    def find(self, node):
        if self.parents[node] == node:
            return node
        ## Path Compression
        self.parents[node] = self.find(self.parents[node])
        return self.parents[node]
    
    def union(self, x, y):
        a = self.find(x)
        b = self.find(y)
        self.parents[b] = a

ds_obj = disjoint_set(5)
ds_obj.union(1, 2)
ds_obj.union(4, 0)
ds_obj.union(0, 3)
ds_obj.union(2, 3)
print(ds_obj.find(3))
ds_obj.parents
```

```python
class disjoint_set:
    def __init__(self, n):
        self.parents = [i for i in range(n)]
        self.rank = {i:0 for i in range(n)}

    def find(self, x):
        if self.parents[x] == x:
            return x
        self.parents[x] = self.find(self.parents[x])
        return self.parents[x]
    
    def union(self, x, y):
        a = self.find(x)
        b = self.find(y)
        if self.rank[b] > self.rank[a]:
            a, b = b, a
        self.parents[b] = a
        if self.rank[b] == self.rank[a]:
            self.rank[a] += 1

ds_obj = disjoint_set(5)
ds_obj.union(1, 2)
ds_obj.union(4, 0)
ds_obj.union(0, 3)
ds_obj.union(2, 3)
print(ds_obj.find(3))
ds_obj.parents

        
```

## **BFS**

Graph traversal algorithm that starts with the source node and traverse level by level. First visits all the neighbouring node of each node and then go to next level. Uses queue for implementation.

- Time complexity: O(V + E)
- Space complexity: O(V + E)

Used for:
1. Shortest path (unweighted)
2. level order traversal
3. Multi-source BFS

```python
from  collections import deque

def bfs(n, graph):
    q = deque([0])
    path = list()
    visited =[0]*n
    visited[0] = True
    while len(q):
        node = q.popleft()
        path.append(node)
        for nb in graph[node]:
            if not visited[nb]:
                q.append(nb)
                visited[nb] = True 
    return path

n = 5
graph = [[1, 3], [0, 3], [3], [0, 2, 4], [3]]
bfs(n, graph)
```
