## Union Find (Disjoint Set)

Union-Find, also known as Disjoint Set Union, is a data structure used to tracks and merge non-overlapping (disjoint) sets.

It supports 2 operation:
1. **Find(x)**: Finds which set element x belongs to.
2. **Union(x, y)**: Merges the set containing x and y.

#### Optimizations:
1. **Path Compression (in find)**: While finding the root node of an element, point every node directly to root. Flattens the tree over time. Reduces the search time from o(n) to o(1).
2. **Union by rank/size (in union)**:

**Uses**:
1. Number of connnected components
2. Detect cycle in an undirected graph

```python
## Template:
class DisjointSet:
    def __init__(self, n):
        self.node = [i for i in range(n)]
    
    def find(self, x):
        if self.node[x] == x:
            return x
        ## Path compression
        self.node[x] = self.find(self.node[x])
        return self.node[x]
    
    def union(self, x, y):
        x = self.find(x)
        y = self.find(y)
        self.node[y] = x

obj = DisjointSet(8)
obj.union(2,3)
obj.union(1,4)
obj.union(1,3)
obj.find(2)
```

```python
## Rank
class DisjointSet:
    def __init__(self, n):
        self.parent = [i for i in range(n)]
        self.rank = [0]*n
        self.components = n

    def find(self, x):
        if self.parent[x] == x:
            return x
        self.parent[x] = self.find(self.parent[x])
        return self.parent[x]
    
    def connected(self, x, y):
        return self.find(x) == self.find(y)
    
    def union(self, x, y):
        px = self.find(x)
        py = self.find(y)
        if px == py:
            return False
        if self.rank[px] < self.rank[py]:
            px, py = py, px
        self.parent[py] = px
        if self.rank[px] == self.rank[py]:
            self.rank[px] += 1
        self.components -= 1
        return True

dsobj = DisjointSet(5)
dsobj.union(1,2)
dsobj.union(3,4)
dsobj.union(0,4)
dsobj.find(3)
```

## Topological Sort (Directed Acyclic Graph)

Topological Sort orders the vertices of a graph in such a way that for every directed edge u->v, u comes before v. If the graph has a cycle, no valid topological sort exists.

Two Standard Approaches
1. A. Kahn's Algorithm (BFS-based, uses in-degree)
2. B. DFS-based (uses finish-time / postorder)

```python
##  A. Kahn's Algorithm (BFS-based, uses in-degree)

from collections import defaultdict, deque

def topo_sort(n , edges):
    order = []
    indegree = [0]* n
    graph = defaultdict(list)
    for u, v in edges:
        indegree[v] += 1
        graph[u].append(v)

    q = deque([i for i in range(n) if indegree[i] == 0])
    while len(q):
        u = q.popleft()
        order.append(u)
        for v in graph[u]:
            indegree[v] -= 1
            if indegree[v] == 0:
                q.append(v)

    return order

n = 5
edges = [[0, 4], [4, 1], [1, 3], [0, 2], [2, 1]]
topo_sort(n ,edges)
```
